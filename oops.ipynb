```python
#no constructor
class Area:
    len = 0
    bth = 0
    def calc(self):
        ans=self.len*self.bth
        return ans
```

```python
# constructor
class Area2:
    def _init_(self, len, bth):
        self.l = len
        self.b = bth

    def calc(self):
        ans = self.l * self.b
        return ans
```

```python
o1=Area()
o1.len=5
o1.bth=9
res=o1.calc()
print(o1.len,o1.bth,res)
```

```python
o2=Area()
o2.len=6
o2.bth=9
res=o2.calc()
print(o2.len,o2.bth,res)
```

```python
o3=Area()
o3.len=6
o3.bth=84
res=o3.calc()
print(o3.len,o3.bth,res)
```

```python

o4 = Area2(15, 5)
print(o4.l, o4.b, o4.calc())

o5 = Area2(10, 13)
res = o5.calc()
print(o5.l, o5.b, res)
```

```python
class Dog:
    def __init__(self, breed, color, age):
        self.breed = breed
        self.color = color
        self.age = age

    def show(self):
        print('the dog details')
        print('Breed:', self.breed)
        print('Color:', self.color)
        print('Age:', self.age)
        
```

```python


bruno = Dog('Golden Retriever', 'Golden', 3)
buddy = Dog('Dalmatian', 'Black & White', 5)
sammy = Dog('Bulldog', 'Black', 6)
atex = Dog('German Shepherd', 'Brown', 4)
bruno.show()
buddy.show()
sammy.show()
atex.show()
```

```python
doglist = [bruno,sammy,atex,buddy]
#display details of the eldest dogs using logic
age = 0
eldest_dog = None
for dog in doglist:
    if age <= dog.age:
        age = dog.age
        eldest_dog = dog
print(eldest_dog.show())    
```

```python
class Pokemon:
    def __init__(self,name,type='normal',hp=10,att=10,dfn=10):
        self.name = name
        self.type = type
        self.hp = hp
        self.atk=att
        self.dfn=dfn
    def stats(self):
        print("pokemon details")
        print(f'Name:{self.name}')
        print(f'Type:{self.type}')
        print(f'Hp:{self.hp}')
        print(f'Attack:{self.atk}')
        print(f'Dfnense:{self.dfn}')

    def overall(self):
        return self.hp + self.atk + self.dfn
    def is_powerfull(self):
        if self.overall()>100:
            return True
        return False
        
        
```

```python

pikachu = Pokemon("pikachu", "electric", 35, 55, 40)
snorlax = Pokemon("snorlax", "normal", 160, 110, 65)
charizard = Pokemon("charizard", "fire/flying", 78, 84, 78)
bulbasaur = Pokemon("bulbasaur", "grass/poison", 45, 49, 49)
ratata = Pokemon("rattata")
pokemon = [pikachu, snorlax, charizard, bulbasaur, ratata]


print("\nStats:")
for poke in pokemon:
    poke.stats()
    print(f'Overall score: {poke.overall()}')
    print(f'Is powerful: {poke.is_powerfull()}')
    print('-----------------------')
```

```python
class Calc:
    def run(self, num1, operator, num2):
        match operator:
            case '+': return self.add(num1, num2)
            case '-': return self.sub(num1, num2)
            case '*': return self.mul(num1, num2)
            case '/':return self.div(num1, num2)
        

    def add(self, a, b): return a + b

    def sub(self, a, b): return a - b

    def mul(self, a, b):return a * b

    def div(self, a, b):
        if b == 0:
            return "Can't divide by 0"
        return a / b

```

```python
c = Calc()

while True:
    a = int(input("Enter first num"))
    b = int(input("Enter next num: "))
    op = input("Enter next (+,-,*,/): ")
    
    ans = c.run(a, op, b)
    print(ans) 
    if input('continue') != 'y':
       
        break
```

```python
class SI:
    def __init__(self,p,r,t):
        self.p = p
        self.r = r
        self.t =t
    def calc(self):
        ans=self.p*self.r*self.t/100
        return ans    
```

```python
class CI(SI):
    def calc(self):
        a=self.p*(1+self.r/100)**self.t
        return a
```

```python
class Player:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    # define how to compare two players using '>'
    def __gt__(self, other):
        return self.score > other.score


p1 = Player('Sam', 20)
p2 = Player('Ram', 15)

if p1 > p2:
    print(f'{p1.name} is better')
else:
    print(f'{p2.name} is better')
```
