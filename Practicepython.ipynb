```python
def is_valid_parentheses(s):
    stack = []
    # Mapping of closing to opening brackets
    bracket_map = {')': '(', '}': '{', ']': '['}
    
    for char in s:
        if char in bracket_map.values():  # opening brackets
            stack.append(char)
        elif char in bracket_map:         # closing brackets
            if stack and stack[-1] == bracket_map[char]:
                stack.pop()
            else:
                return False
    return len(stack) == 0

# Input parentheses string
s = input("Enter parentheses string: ")
print(is_valid_parentheses(s))
```

```python
def power(a, b):
    if b == 0:
        return 1
    half = power(a, b // 2)
    if b % 2 == 0:
        return half * half
    else:
        if b > 0:
            return half * half * a
        else:
            return (half * half) / a

a = 2
b = 10
print(power(a,b))
```

```python
#find missing number
def find_missing(nums):
    n = len(nums) + 1
    expected_sum = n * (n+1) // 2
    actual_sum = sum(nums)
    return expected_sum - actual_sum
nums = [1,2,4,5]
print("missing number is:",find_missing(nums))
```

```python
#bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j],arr[j + 1] = arr[j + 1],arr[j]
                return arr
nums = [5,1,4,2,8]
print("sorted list:",bubble_sort(nums))
            
```

```python
#transpose matrix
def transpose_matrix(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    result = [[0 for _ in range(rows)] for _ in range(cols)]

    for i in range(rows):
        for j in range(cols):
            result[j][i] = matrix[i][j]
    return result
matrix = [
    [1,2,3],
    [4,5,6]
] 

print("transpose:")
for row in transpose_matrix(matrix):
    print(row)
```

```python
def second_largest(nums):
    unique_nums = list(set(nums))
    if len(unique_nums) < 2:
        return "no second largest element"
    unique_nums.sort()
    return unique_nums[-2]

nums = [4,1,2,4,3,2]
print("second largest number is:",second_largest(nums))
```

```python
#anagram checker
def is_anagram(s1,s2):
    s1 = s1.replace(" ", "").lower()
    s2 = s2.replace(" ", "").lower()
    return sorted(s1) == sorted(s2)

s1 = "listen"
s2 = "silent"
print(is_anagram(s1,s2))
```

```python
#armstrong number check
def is_armstrong(n):
    original = n
    sum_cubes = 0
    while n > 0:
        digit = n % 10
        sum_cubes += digit ** 3
        n //= 10
    return sum_cubes == original

n = 153
print(f"{n} is armstrong:",is_armstrong(n))

n = 123
print(f"{n} is armstrong:",is_armstrong(n))
```

```python
#substring count
def count_substring(text,pattern):
    count = 0
    i = 0
    while i <= len(text) - len(pattern):
        if text[i:i+len(pattern)] == pattern:
            count += 1
        i += 1
    return count

text = "abababa"
pattern = "aba"
print("occurrences:",count_substring(text,pattern))    
```

```python
#merge two sorted list
def merge_sorted_lists(list1, list2):
    merged = []
    i = j = 0
    n1, n2 = len(list1), len(list2)

    while i < n1 and j < n2:
        if list1[i] < list2[j]:
            merged.append(list1[i])
            i += 1
        else:
            merged.append(list2[j])
            j += 1

    # Add remaining elements
    while i < n1:
        merged.append(list1[i])
        i += 1
    while j < n2:
        merged.append(list2[j])
        j += 1

    return merged

list1 = [1, 3, 5]
list2 = [2, 4, 6]
print("Merged list:", merge_sorted_lists(list1, list2))
```

```python
def unique_ordered(lst):
    seen = set()
    unique_lst = []
    for item in lst:
        if item not in seen:
            seen.add(item)
            unique_lst.append(item)
    return unique_lst

nums = [1,2,2,3,4,3]      
print("unique elements (ordered):",unique_ordered(nums))  
```
