```python
movies=['og','pushpa','kannapa','jinna','journey','little hearts','court','luckybasker','thamudu','santhosam','kalisundam raa','nuvu nakinachav','baby','7g bhrudhavan colony','remo','joe','orange','arjun reddy','rx100']
movies_caps = []
for movie in movies:
    name = movie.upper()
    movies_caps.append(name)
print(movies)
print(movies_caps)    
```

```python
movies_caps = [movie.upper() for movie in movies]
print(movies)
print(movies_caps)
```

```python
#filter the movies the contain the char 'o' and store in new list
movies_with_o = []
for item in movies:
    if 'o' in item or 'O' in item:
        movies_with_o.append(item)
print(movies)
print(movies_with_o)        
```

```python
movies_with_e = [nm for nm in movies if 'o' in nm or 'O' in nm]
print(movies_with_o)
```
