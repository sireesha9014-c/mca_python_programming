```python
# a simple contact book using sttorage(dictionary)
contact = {
    'emergency' : 108
}
print("Contact Book")
while True :
    print("Search contact")
    name = input("Name ")
    if name in contact:
        print("contact found")
        print(name,contact.get(name))
    else:
        print(" contact not found")
        ch = input("do you want to add it ? y/n:")
        if ch == 'Y':
            num = input("Enter {name}'s number:")
            contact[name] = num
            print(" success! contact added")
        else:
            print("okay")   
    quit = input("do you want to continue? Y/n")
    if quit != 'Y':
        print("Bye")
        break
```
