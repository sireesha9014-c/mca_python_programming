```python
x = np.random.randint(1,100,(10,10))
print(x)
```

```python
print("first row", x[0])
print("last row", x[-1])
print("second row", x[1])
print('first 5 element of first row', x[0,:5])
print('last 5 element of first row', x[0,-3:])
print('first 3 element of last row', x[-1,:3])
print('third row, all except first 2 and last 2', x[2,2:-2])
print('top 3 rows\n', x[:3])
print('last 3 row\n',x[-3:])
print('all rows except first and last\n', x[1:-1])
```

```python
print("1st column", x[:,0])
print("last column of x" , x[:,-1])
print('5th column of x', x[:,4])
print('first 3 columns' , x[:, :3])
print('last 4 columns', x[:, -4:])
print('all the columns except first and last\n', x[:,1:-1])
```

```python
print('all the odd columns', x[:,1::2])
```

BOOLEAN INDEXING AKA FILTERING

```python
x = np.random.randint(1,10, (4,10))
print(x)
```

```python
x>5  # boolean filter -a matrix of true and false, with same dimension as x matrix
```

```python
x[x>3] #using boolean filter to extract values that are true
```

```python
filter1 = x > 3
filter2 = x < 7
# extract using more than 1 filter
x[filter1 & filter2]  # or x[(x>3) & (x<7)]
```
