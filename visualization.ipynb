```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Segoe UI Emoji'

events = ['start', 'Hello\nworld', 'Datatypes', 'Control Flow', 'Data Structures', 
          'Functions', 'OOP', 'Data Analysis', 'Web Dev', 'Project']
weeks = np.arange(len(events))
confidence = [7, 10, 8, 6, 7, 8, 6, 3, 2, 5]

plt.figure(figsize=(10, 6))

plt.plot(weeks, confidence, marker='o', linestyle='--', markersize=8, 
         label='Confidence Level')

plt.annotate('Peak of confidence', xy=(1, 10), xytext=(3, 9),
             arrowprops=dict(facecolor='black', shrink=0.05, width=2),
             ha='center')

plt.axvspan(7, 10, color='yellow', alpha=0.3, hatch='o', label='Challenging part')
plt.axvspan(1, 3, color='blue', alpha=0.1, hatch='//')

plt.xticks(weeks, events, rotation=45)
plt.xlabel('Learning Milestones')
plt.ylabel('Confidence Level (1 - 10)')
plt.grid(True)
plt.ylim(0, 11)
plt.tight_layout()
plt.legend()
plt.show()
```
