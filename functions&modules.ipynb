```python
#default args
def cleanText(data, to_remove="@s#%"):
    newdata = data #make a copy of data
    for char in to_remove: #loop of each char to remove
        newdata = newdata.replace(char,'') 
    return newdata

cText = cleanText("He@#llo@")
print(cText)
cText2 = cleanText("H@!()#!)@i",to_remove='@!#()')   
print(cText2) 
```

### variable argument is python
 a function can have variable arguments in python using symbol before a paramter in the function definition this allows the function to accept any number of positional arguments as a tuple.

```python
def multiplier (*numbers):
    result = 1
    for value in numbers:
        result *= value
    return result
print(multiplier())
print(multiplier(2,3,4))
print(multiplier(2,3,4,3,5,6,1,2,2,5,56,2))
```

```python
def operate(*args,operate="*"):
    if operate == '*':
        result=1
        for num in args:
            result*=num
        return result
    elif operate == '+' :
        result = 0
        for num in args:
            result+=num
        return result
    else:
        return "Wrong operator give ,please provide + or *"
print(operate(1,2,3,4))  #default multiply values
print(operate(1,2,3,4,operate='+'))     # operator is +,so sum of value  
print(operate(1,2,3,4,operate='-'))  # wrong operator
```
