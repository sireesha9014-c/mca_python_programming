# Generators or lazy functions
special type of function that will not evaluate and return all values at once.inside it will return one value at a time,and only when requested

```python
for i in range(1,6): print(i)
print(list(range(1,10)))
```

```python
# create a generator (use yield insted of return)
def squared(data):
    for i in data:
        yield i**2 # yield stores all result in memory to be 
                   # used later and dont return a single value
#squared(data=[1,2,3,4,5,6,7,8])   # this  will not give you result


# this is how we use generator(getting 1 value at a time)
for num in squared([2,3,3,5,6,7]):
    print("squared",num)

    #or
print(list(squared([22,33,55])))    # this calcuates and convert all values
```
