```python
class Student:
    def _init_(self,name,marks=[]):
        self.name=name
        self.marks=marks
    def total_marks(self):
        return sum(self.marks)
    def _gt_(self,other):
        if isinstance(other,Student):
            return self.total_marks()>other.total_marks()
        elif isinstance(other,int):
            return self.total_marks()>other
    def _repr_(self):
        return f'Student({self.name},{self.total_marks()})'
def is_passed(marks):
    return marks>200



students=[
    Student("charishma",[67,78,98,78]),
    Student("anupama",[86,75,48,68]),
    Student("vishwas",[84,78,45,43]),
    Student("vijay",[34,56,23,67]),
]
list(filter(is_passed,students))
```
