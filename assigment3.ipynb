```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)
```

```python
print(factorial(5))
```

```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(10))
```

```python
class Rectangle:
    def __init__(self, length=0, width=0):
        self.length = length
        self.width = width

    def set_dimensions(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


# Test code should be OUTSIDE the class
r = Rectangle()
r.set_dimensions(5, 3)
print("Area:", r.area())           # 15
print("Perimeter:", r.perimeter()) # 16
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks  # list of marks for 3 subjects

    def total_marks(self):
        return sum(self.marks)

    def average_marks(self):
        return self.total_marks() / len(self.marks)

s = Student("Alice", [85, 90, 78])
print("Total:", s.total_marks())     # 253
print("Average:", s.average_marks()) # 84.33
```

```python
class Temperature:
    def __init__(self, celsius=0):
        self.celsius = celsius

    def set_celsius(self, celsius):
        self.celsius = celsius

    def to_fahrenheit(self):
        return (self.celsius * 9/5) + 32

    def to_kelvin(self):
        return self.celsius + 273.15

t = Temperature()
t.set_celsius(25)
print("Fahrenheit:", t.to_fahrenheit()) # 77.0
print("Kelvin:", t.to_kelvin())         # 298.15
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = []  # list of dictionaries [{'name': 'item1', 'price': 100}]

    def add_item(self, name, price):
        self.items.append({'name': name, 'price': price})

    def remove_item(self, name):
        self.items = [item for item in self.items if item['name'] != name]

    def total_price(self):
        return sum(item['price'] for item in self.items)

cart = ShoppingCart()
cart.add_item("Apple", 2)
cart.add_item("Banana", 1)
cart.remove_item("Apple")
print("Total:", cart.total_price())  # 1
```

```python
class Employee:
    def __init__(self, name, emp_id, salary):
        self.name = name
        self.emp_id = emp_id
        self.salary = salary

    def give_raise(self, percentage):
        self.salary *= (1 + percentage / 100)


e = Employee("John", 101, 50000)
e.give_raise(10)
print("New Salary:", e.salary)  # 55000.0
```
